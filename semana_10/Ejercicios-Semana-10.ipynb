### Ejercicio 2
Complete una clase nueva llamada *Capacitor* que permita definir un objeto capacitavo de la forma:

    c1 = Capacitor(3e-12)      # Un capacitor de 3 picoFaradios
    
Valide que el valor del capacitor solo pueda ser un valor numérico (ya sea int o float).

Considere ajustar la forma como un objeto de la clase Capacitor se comportará ante las operaciones. Si se desea calcular la capacitancia equivalente en paralelo de 2 capacitores se podrá aplicar el operador // y si desea calcular la capacitancia en serie de 2 capacitores se podrá aplicar el operador +. De tal forma que puede tener:

    c2 = Capacitor(6e-12)    # Un capacitor de 6 picoFaradios
    c3 = Capacitor(8e-12)    # Un capacitor de 8 picoFaradios
    
    result = c1 + c2 + c3    # result sera igual a 1.6 picofaradios
    result = c1 // c2 // c3  # result sera igual a 17 picofaradios
    
La operación *c1 // c2* llama al método privado *c1.\__floordiv\__(c2)*. 

Por ejemplo, la suma *c1 + c2* esta definida como *c1.\__add\__(c2)* y se encuentra codificada para su referencia.
